```python
import pandas as pd
import numpy as np
```

Creamos una función que cuenta los valores nulos por columna junto con su porcentaje respectivo:

```python
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # número total de valores nulos
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)  # porcentaje de valores nulos
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  # uniendo las dos columnas anteriores
    return missing_data.reset_index() # devuelve el dataframe
```

Imputamos el desempeño en los datos nulos de la columna "result_desempeno":

```python
import pandas as pd

# Definir las condiciones y valores
condiciones = [
    df2016_esp_lim['result_puntaje'].between(0, 124),
    df2016_esp_lim['result_puntaje'].between(125, 154),
    df2016_esp_lim['result_puntaje'].between(155, 199),
    df2016_esp_lim['result_puntaje'].between(200, 300)
]

valores = [1, 2, 3, 4]

# Aplicar la imputación solo a los valores vacíos
df2016_esp_lim['result_desempeno'] = df2016_esp_lim['result_desempeno'].mask(
    df2016_esp_lim['result_desempeno'].isna(),
    pd.Series(np.select(condiciones, valores, default=np.nan))
)
```

```python
# Eliminar datos vacíos de 'result_desempeno'
df2016_esp_lim = df2016_esp_lim.dropna(subset=['result_desempeno'])
```

```python
missing_check(df2016_esp_lim)
```

```python
import pandas as pd

def comparar_columnas(df1, df2, nombre_df1="DataFrame 1", nombre_df2="DataFrame 2"):
    """
    Compara las columnas comunes y no comunes entre dos DataFrames
    """
    columnas_df1 = set(df1.columns)
    columnas_df2 = set(df2.columns)
    
    print(f"=== COMPARACIÓN DE COLUMNAS ===")
    print(f"Columnas en {nombre_df1}: {len(columnas_df1)}")
    print(f"Columnas en {nombre_df2}: {len(columnas_df2)}")
    print("-" * 40)
    
    # Columnas comunes
    comunes = columnas_df1.intersection(columnas_df2)
    print(f"Columnas COMUNES ({len(comunes)}):")
    print(sorted(comunes))
    print("-" * 40)
    
    # Columnas solo en df1
    solo_df1 = columnas_df1 - columnas_df2
    print(f"Columnas solo en {nombre_df1} ({len(solo_df1)}):")
    print(sorted(solo_df1))
    print("-" * 40)
    
    # Columnas solo en df2
    solo_df2 = columnas_df2 - columnas_df1
    print(f"Columnas solo en {nombre_df2} ({len(solo_df2)}):")
    print(sorted(solo_df2))
    
    return {
        'comunes': sorted(comunes),
        'solo_df1': sorted(solo_df1),
        'solo_df2': sorted(solo_df2)
    }

# Uso
comparar_columnas(df2021_esp_lim, df2019_esp_lim, "DataFrame 2021", "DataFrame 2019")
```
