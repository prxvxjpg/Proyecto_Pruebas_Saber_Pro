Ahora, creamos una función que cuenta los valores nulos por columna junto con su porcentaje respectivo:

```python
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # número total de valores nulos
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)  # porcentaje de valores nulos
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  # uniendo las dos columnas anteriores
    return missing_data.reset_index() # devuelve el dataframe

missing_check(df) # Corremos la función para ver los valores nulos
```

Creamos una función que asigne valores específicos a los valores nulos en diferentes columnas del DataFrame:

```python
def reemplazar_nan_personalizado(df, columnas_valores):
    """
    Reemplaza valores NaN en columnas específicas con valores personalizados
    
    Parámetros:
    df (pd.DataFrame): DataFrame original
    columnas_valores (dict): Diccionario con {columna: valor_para_nan}
    
    Returns:
    pd.DataFrame: DataFrame con los NaN reemplazados
    """
    df_temp = df.copy()
    
    for columna, valor in columnas_valores.items():
        if columna in df_temp.columns:
            df_temp[columna] = df_temp[columna].fillna(valor)
        else:
            print(f"Advertencia: La columna '{columna}' no existe en el DataFrame")
    
    return df_temp
```

**Manejo de Valores en la Variable de Estrato Socioeconómico**

De acuerdo con la documentación del dataset, los valores nulos y los registros etiquetados como 'Sin Estrato' en la columna `fami_estratovivienda` probablemente corresponden a **hogares ubicados en zonas rurales donde no existe estratificación socioeconómica formal**.

**Estrategia de Imputación:**
- Asignaremos el **valor numérico 0** a los datos nulos
- Convertiremos los registros 'Sin Estrato' a **"Estrato 0"**

**Fundamentación de la Decisión:**
1. **Contexto Legal vs. Analítico:** 
   - Si bien el "Estrato 0" no existe en la legislación colombiana, creamos esta categoría analítica para representar una realidad específica: zonas rurales marginadas sin clasificación oficial.

2. **Criterio Socioeconómico:**
   - Asignamos el valor más bajo (0) debido a la correlación documentada entre la falta de estratificación formal y condiciones de vulnerabilidad socioeconómica en estas zonas.

3. **Preservación de Datos:**
   - La principal motivación es **evitar la pérdida de información valiosa** manteniendo estos registros en el análisis.

**Justificación Técnica para Machine Learning:**
Los algoritmos de ML requieren comúnmente:
- ✅ Variables numéricas completas
- ✅ Ausencia de valores nulos
- ✅ Representación consistente de categorías

Esta transformación nos permite:
- Conservar todos los registros para análisis posteriores
- Mantener una escala ordinal coherente (0-6)
- Facilitar el procesamiento por modelos predictivos


```python
# Primero manejar "Sin estrato" y luego los valores vacíos
df['fami_estratovivienda'] = df['fami_estratovivienda'].replace('Sin estrato', 'Estrato 0')

# Extraer el número y manejar valores vacíos
df['fami_estratovivienda'] = (
    df['fami_estratovivienda']
    .str.extract('(\d)')
    .astype(float)  # Usar float para permitir NaN
    .fillna(0)      # Llenar vacíos con 0 (rural/sin estrato)
    .astype(int)    # Convertir a entero
)
```

Función para limpiar texto en columnas específicas con opciones personalizables:

```python
import re

def limpiar_texto_personalizable(df, columnas_texto, 
                                eliminar_espacios=True,
                                a_minusculas=True, 
                                eliminar_puntuacion=True,
                                preservar_numeros=True,
                                ):
    """
    Función altamente personalizable para limpieza de texto
    
    Parámetros:
    - eliminar_espacios: Elimina espacios extras
    - a_minusculas: Convierte a minúsculas
    - eliminar_puntuacion: Elimina signos de puntuación
    - preservar_numeros: Mantiene números
    """
    
    df_clean = df.copy()
    
    for columna in columnas_texto:
        if columna in df_clean.columns:
            df_clean[columna] = df_clean[columna].astype(str)
            
            # Aplicar transformaciones según parámetros
            if eliminar_espacios:
                df_clean[columna] = df_clean[columna].str.strip()
                df_clean[columna] = df_clean[columna].str.replace(r'\s+', ' ', regex=True)
            
            if a_minusculas:
                df_clean[columna] = df_clean[columna].str.lower()
            

            
            # Eliminar puntuación pero mantener diacríticos y números
            if eliminar_puntuacion:
                if preservar_numeros:
                    # Mantiene letras (con diacríticos), números y espacios
                    df_clean[columna] = df_clean[columna].str.replace(
                        r'[^\w\sáéíóúüñÁÉÍÓÚÜÑ0-9]', '', regex=True
                    )
                else:
                    # Mantiene solo letras (con diacríticos) y espacios
                    df_clean[columna] = df_clean[columna].str.replace(
                        r'[^\w\sáéíóúüñÁÉÍÓÚÜÑ]', '', regex=True
                    )
    
    return df_clean
df = limpiar_texto_personalizable(df, ["estu_prgm_academico"],a_minusculas=False) # Mantenemos mayusculas con a_minusculas=False
df.head()

# Cantidad de programas académicos únicos
print("Cantidad de programas académicos únicos:")
print(len(df["estu_prgm_academico"].unique()))
```

Función para reemplazar cadenas por otras cadenas en una columna de un dataframe:

```python
# Función para reemplazar cadenas por otras cadenas en una columna de un dataframe.
def reemplazar_cadenas(df, columna, mapeo_reemplazos, caso_sensitivo=True, regex=False):
    """
    Función avanzada para reemplazar múltiples cadenas
    
    Parámetros:
    df: DataFrame
    columna: nombre de la columna
    mapeo_reemplazos: diccionario {cadena_original: cadena_nueva}
    caso_sensitivo: True para distinguir mayúsculas/minúsculas
    regex: True si usas expresiones regulares
    
    Retorna:
    DataFrame modificado
    """
    df_modificado = df.copy()
    
    # Convertir a string si no lo es
    df_modificado[columna] = df_modificado[columna].astype(str)
    
    for original, nuevo in mapeo_reemplazos.items():
        df_modificado[columna] = df_modificado[columna].str.replace(
            original, nuevo, case=caso_sensitivo, regex=regex
        )
    
    return df_modificado
```

Creamos una función que compara dos columnas y devuelve un DataFrame con las filas donde los valores de ambas columnas coinciden:

```python
def comparar_columnas(df1, col1, df2, col2):
    """
    Compara dos columnas de diferentes DataFrames y devuelve los elementos no comunes
    """
    # Convertir a conjuntos (eliminando NaN si los hay)
    set1 = set(df1[col1].dropna())
    set2 = set(df2[col2].dropna())
    
    # Calcular las diferencias
    solo_en_df1 = list(set1 - set2)
    solo_en_df2 = list(set2 - set1)
    comunes = list(set1 & set2)
    no_comunes = list(set1.symmetric_difference(set2))
    
    resultado = {
        'solo_en_primer_df': sorted(solo_en_df1),
        'solo_en_segundo_df': sorted(solo_en_df2),
        'comunes': sorted(comunes),
        'no_comunes': sorted(no_comunes),  # lista de todos los valores no comunes
        'estadisticas': {
            'total_primer_df': len(set1),
            'total_segundo_df': len(set2),
            'total_comunes': len(comunes),
            'total_no_comunes': len(no_comunes)
        }
    }
    
    # Imprimir reporte detallado
    print("=" * 50)
    print(f"COMPARACIÓN: {col1} vs {col2}")
    print("=" * 50)
    print(f"Elementos únicos en {col1}: {len(set1)}")
    print(f"Elementos únicos en {col2}: {len(set2)}")
    print(f"Elementos comunes: {len(comunes)}")
    print(f"Elementos NO comunes: {len(no_comunes)}")
    print(f"Elementos solo en primer DataFrame: {len(solo_en_df1)}")
    print(f"Elementos solo en segundo DataFrame: {len(solo_en_df2)}")
    
    print(f"\nVALORES NO COMUNES (diferencia simétrica):")
    print(f"No comunes: {sorted(no_comunes)}")
    
    print(f"\nDETALLE:")
    print(f"Solo en {col1}: {sorted(solo_en_df1)}")
    print(f"Solo en {col2}: {sorted(solo_en_df2)}")
    print(f"Comunes: {sorted(comunes)}")
    
    return resultado
```

El siguiente código toma la fecha de nacimiento del estudiante en la columna `estu_fechanacimiento`, la convierte en un formato de fecha estructurado y extrae el año de nacimiento para crear una nueva columna llamada `ano_nacimiento` la cual se va a colocar al lado derecho de la columna `estu_fechanacimiento`.

(Tenga en cuenta que el siguiente código va a eliminar filas que no se puedan parsear correctamente, es decir, que no se pueda extraer el año de nacimiento a partir de la fecha de nacimiento proporcionada.)

```python
# Primero guardamos un copia del DataFrame original antes de modificar
df_original = df.copy()

# Comprobar columna de referencia
if 'estu_fechanacimiento' not in df.columns:
    raise KeyError("No se encontró la columna 'estu_fechanacimiento' en el DataFrame.")

# 1) Intento principal: parsear (convertir una cadena de texto en un formato de fecha estructurado) asumiendo dayfirst=True (DD/MM/YYYY)
s = pd.to_datetime(df['estu_fechanacimiento'], dayfirst=True, errors='coerce', infer_datetime_format=True)

# 2) Para los que quedaron NaT, intentar parsear asumiendo dayfirst=False (YYYY-MM-DD u otros)
mask_na = s.isna()
if mask_na.any():
    s2 = pd.to_datetime(df.loc[mask_na, 'estu_fechanacimiento'], dayfirst=False, errors='coerce', infer_datetime_format=True)
    s.loc[mask_na] = s2

# Preparar la serie final de años (nullable integer)
years = s.dt.year.astype('Int64')

# 3) Obtener la posición de la columna de referencia (justo a la derecha)
posicion = df.columns.get_loc('estu_fechanacimiento') + 1

# 4) Insertar la nueva columna en la posición deseada
df.insert(posicion, 'ano_nacimiento', years)

# 5) Informes rápidos antes de filtrar
n_total = len(df)
n_parsed = df['ano_nacimiento'].notna().sum()
n_missing = df['ano_nacimiento'].isna().sum()
print(f"Filas totales: {n_total:,}")
print(f"Años extraídos: {n_parsed:,}")
print(f"No parseables / NA: {n_missing:,}")

# Mostrar algunos ejemplos (hasta 8) de valores no parseables en la columna original
if n_missing > 0:
    ejemplos = df.loc[df['ano_nacimiento'].isna(), 'estu_fechanacimiento'].drop_duplicates().head(8).to_list()
    print('\nEjemplos de valores no parseables (hasta 8):')
    print(ejemplos)

# 6) Eliminar filas donde ano_nacimiento es NA (no parseables)
n_before = len(df)
df = df[df['ano_nacimiento'].notna()].copy()
n_after = len(df)
n_dropped = n_before - n_after
print(f"\nFilas antes: {n_before:,} | Filas después: {n_after:,} | Filas eliminadas (no parseables): {n_dropped:,}")

# (Opcional) resetear índice
df.reset_index(drop=True, inplace=True)

# Mostrar ejemplo final (últimas 10 filas)
print('\nEjemplo final (últimas 10 filas):')
display(df[['estu_fechanacimiento','ano_nacimiento']].tail(10))
```
